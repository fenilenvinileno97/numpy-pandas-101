```python
dict_1 = {
    'A':['A0', 'A1', 'A2', 'A3'], 
    'B':['B0', 'B1', 'B2', 'B3'], 
    'C':['C0', 'C1', 'C2', 'C3'], 
    'D':['D0', 'D1', 'D2', 'D3']
    }
```

```python
df1 = pd.DataFrame(dict_1)
```

```python
dict_2 = {
    'A': ['A4', 'A5', 'A6', 'A7'],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'C': ['C4', 'C5', 'C6', 'C7'],
    'D': ['D4', 'D5', 'D6', 'D7']
}
```

```python
df2 = pd.DataFrame(dict_2)
```

Through `.concat()` method, we can merge dataframes through an specified axis. In this case, *axis=0* means that we'll merge by rows. *ignore_index* is a property which includes or does not include the index list.

```python
pd.concat([df1, df2], axis=0, ignore_index=True)
```

```python
pd.concat([df1, df2], axis=1, ignore_index=True)
```

#### Merge


```python
left = pd.DataFrame({
    'A':['A0', 'A1', 'A2', 'A3'], 
    'B':['B0', 'B1', 'B2', 'B3'], 
    'key': ['k0','k1','k2','k3']
    })
left
```

```python
right = pd.DataFrame(
    {'C':['C0', 'C1', 'C2', 'C3'], 
     'D':['D0', 'D1', 'D2', 'D3'],
     'key_2': ['k0', 'k2', 'k3', 'k4']}
    )
right
```

```python
left.merge(right, left_on='key', right_on='key_2', how='left')
```

`.join()` concatenates two dataframes through indexes.

```python
right.join(left)
```

```python
right.join(left, how='inner')
```
