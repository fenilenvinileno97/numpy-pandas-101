```python
import pandas as pd
import numpy as np
```

```python
df_books = pd.read_csv('../imgs/bestsellers.csv', sep=',', header=0)
df_books.head(2)
```

```python
def two_times(value):
    return value * 2
```

```python
df_books['Rating_2'] = df_books['User Rating'].apply(two_times)
df_books[['Name', 'Author', 'User Rating', 'Rating_2']].head()
```

```python
df_books['Rating_2'] = df_books['User Rating'].apply(lambda x : x*3)
df_books.head()
```

```python
df_books['Rating_2'] = df_books.apply(lambda x: x['User Rating']*2 if x['Genre'] == 'Fiction' else x['User Rating'], axis=1)
```
